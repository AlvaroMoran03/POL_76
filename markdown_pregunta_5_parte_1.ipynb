$5) Sea \hat{\beta}_n = (X_n'X_n)^{-1} X_n'Y_n$

$Y_n \in \mathbb{R}^{n \times 1}, \quad X_n \in \mathbb{R}^{n \times k}$

Añadiendo una nueva observación: $[Y_{n+1}, X_{n+1}]$

Probar que:
$$
\hat{\beta}_{n+1} = \hat{\beta}_n + \frac{1}{1 + X_{n+1}'(X_n'X_n)^{-1} X_{n+1}}
\cdot (X_n'X_n)^{-1} X_{n+1} \left[Y_{n+1} - X_{n+1}'\hat{\beta}_n\right]
$$
Para ello, partimos de:
$$
\hat{\beta}_{n+1} = (\textcolor{red}{X_{n+1}'X_{n+1}})^{-1} \textcolor{red}{X_{n+1}'Y_{n+1}}
$$


$$
\textcolor{red}{Y_{n+1}} =
\begin{bmatrix}
Y_n \\
Y_{n+1}
\end{bmatrix},
\quad
\textcolor{red}{X_{n+1}} =
\begin{bmatrix}
X_n \\
X_{n+1}'
\end{bmatrix}_{(n+1) \times k}
$$


Como: $\textcolor{red}{X_{n+1}'} = \begin{bmatrix} X_n' \\ X_{n+1} \end{bmatrix}$,
$$
\hat{\beta}_{n+1} =
\left[X_n'X_n + X_{n+1} X_{n+1}' \right]^{-1}
\left(X_n' Y_n + X_{n+1} Y_{n+1} \right)
$$


Sherman-Morrison Formula:
$$
(A + uv')^{-1} = A^{-1} - \frac{1}{1 + v'A^{-1}u} A^{-1}uv'A^{-1}
$$




Definimos:
$$
A = X_n' X_n
$$
$$
u = X_{n+1}
$$
$$
v = X_{n+1}
$$
