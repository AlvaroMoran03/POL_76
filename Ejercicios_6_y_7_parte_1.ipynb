#pregunta 6

Mostrar que  ${[r_{Y, \hat{Y}}]}^2 = R^2  $

Para ello, partimos de:
$$
r_{Y, \hat{Y}} =   \frac{\hat{cov}(Y, \hat{Y})}{{[\hat{var}[Y] \hat{var}[\hat{Y}]]}^{1/2}}  
$$

$$
r_{Y, \hat{Y}} =   \frac{\hat{cov}(\hat{Y} + \hat{e},\hat{Y})}{{[\hat{var}[Y] \hat{var}[\hat{Y}]]}^{1/2}}  
$$

Por construcción, no hay covarianza entre el valor ajustado  $\hat{Y}$ y el residuo  $\hat{e}$

$$
r_{Y, \hat{Y}} =   \frac{\hat{cov}(\hat{Y},\hat{Y}) + 0}{{[\hat{var}[Y] \hat{var}[\hat{Y}]]}^{1/2}}  
$$

$$
r_{Y, \hat{Y}} =   \frac{\hat{var}[\hat{Y}]}{{[\hat{var}[Y] \hat{var}[\hat{Y}]]}^{1/2}}  
$$


$$
r_{Y, \hat{Y}} =   \frac{{[\hat{var}[\hat{Y}]]}^{1/2}}{{[\hat{var}[{Y}]]}^{1/2}}  
$$



Elevando al cuadrado

$$
{[r_{Y, \hat{Y}}]}^2 =   \frac{[\hat{var}[\hat{Y}]]}{[\hat{var}[{Y}]]}
$$

$$
{[r_{Y, \hat{Y}}]}^2 =   \frac{\frac{1}{n-1} \sum {(\hat{y}_i - \bar{y})}^2}{\frac{1}{n-1} \sum {( y_i - \bar{y})}^2}
$$

$$
{[r_{Y, \hat{Y}}]}^2=   \frac{ \sum {(\hat{y}_i - \bar{y})}^2}{ \sum {( y_i - \bar{y})}^2} = R^2
$$

$$
{[r_{Y, \hat{Y}}]}^2 = R^2
$$




#Pregunta 7

Minimizando el estimador de momentos de la suma de cuadrados esperada del error $\hat{S}(b)$

$$
Min \hat{S}(b) =  \sum {(Y_i - b)}^2 
$$

Donde la variable de elección es b

FOC:

$$
\sum -2(Y_i - b) = 0
$$

$$
\sum (Y_i - b)= 0
$$

$$
\sum Y_i= nb
$$

$$
\hat{\beta} = \frac{\sum Y_i}{n} = \bar{Y}
$$

El leave one-out error es:

$$
\tilde{e}_i = (1- h_{ii})^{-1}\hat{e}_i
$$

En este caso:
$\hat{e}_i = Y - \hat{\beta} =   Y - \bar{Y} $
Notemos que $h_{ii} = x'_i(X'X)^{-1}x_i  = 1 \cdot n^{-1} \cdot 1 = \frac{1}{n}$, pues $X$ es un vector de unos.

Reemplazando

$$
\tilde{e}_i = (1- \frac{1}{n})^{-1}(Y - \bar{Y})
$$

$$
\tilde{e}_i = ( \frac{n -1}{n})^{-1}(Y - \bar{Y})
$$


$$
\tilde{e}_i = (\frac{n}{n - 1})(Y - \bar{Y})
$$

Que es lo que queríamos demostrar
