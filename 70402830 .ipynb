For the logistic distribution $\Lambda(x)=(1+\exp (-x))^{-1}$ verify that

 (a) [1 points] $\frac{d}{d x} \Lambda(x)=\Lambda(x)(1-\Lambda(x))$.

_Answer_:
We need to prove the following expression is true:
$$
\frac{d}{dx} \Delta(x) = \Delta(x) \left( 1 - \Delta(x) \right)
$$
Remember:
$$
\Delta(x) = \left( 1 + \exp(-x) \right)^{-1}
$$

   $$
   \frac{d}{dx} \left( 1 + \exp(-x) \right)^{-1} = -1 \cdot \left( 1 + \exp(-x) \right)^{-2} \cdot \left( -\exp(-x) \right)
   $$

   $$
   = \left( 1 + \exp(-x) \right)^{-2} \cdot \left( \exp(-x) + 1 - 1 \right)
   $$
Grouping the terms:
   $$
   = \left( 1 + \exp(-x) \right)^{-2} \cdot \left( \left( \exp(-x) + 1 \right) - 1 \right)
   $$
Rewrite the expression:
   $$
   = \left( 1 + \exp(-x) \right)^{-1} \cdot \left( \left( 1 + \exp(-x) \right)^{-1} \cdot \left( \exp(-x) + 1 \right) - \left( 1 + \exp(-x) \right)^{-1} \right)
   $$
Expand the product:
   $$
   = \left( 1 + \exp(-x) \right)^{-1} \cdot \left( 1 - \left( 1 + \exp(-x) \right)^{-1} \right)
   $$

The resulting expression is:
$$
\frac{d}{dx} \Delta(x) = \Delta(x) \left( 1 - \Delta(x) \right)
$$

E[X \mathbf{1}{\{|x{i}| < c\}} E[e | X]] = 0$

*we get:*

$\tilde{\beta} =

(b) [2 points] Find the asymptotic distribution of $\sqrt{n}(\tilde{\beta}-\beta)$.

_Answer_:

$$
\sqrt{n}(\tilde{\beta} - \beta) = 
\left( \frac{1}{n} \sum_{i=1}^{n} X_{i} X_{i}' \mathbf{1}{\{|x_{i}| < c\}} \right)^{-1} 
\left( \frac{1}{\sqrt{n}} \sum_{i=1}^{n} X_{i} e_{i} \mathbf{1}{\{|x_{i}| < c\}} \right)
$$

Where: 
- \(X_{i} e_{i}\) are i.i.d. with \(E[X_{i} e_{i}] = 0\).  
- \(\Omega = E[XX'e^{2}]\) (finite for CLT).  

Then:

1. The first term converges in probability:  
$$
\left( \frac{1}{n} \sum_{i=1}^{n} X_{i} X_{i}' \mathbf{1}{\{|x_{i}| < c\}} \right)^{-1} 
\xrightarrow{p} \left( E[XX'] \mathbf{1}{\{|x_{i}| < c\}} \right)^{-1} = (Q_{XX1})^{-1}
$$  

2. The second term converges in distribution:  
$$
\frac{1}{\sqrt{n}} \sum_{i=1}^{n} X_{i} e_{i} \mathbf{1}{\{|x_{i}| < c\}} 
\xrightarrow{d} N(0, \Omega)
$$  

By Slutsky's Theorem:
$$
\sqrt{n}(\tilde{\beta} - \beta) \xrightarrow{d} N(0, (Q_{XX1})^{-1} \Omega (Q_{XX1})^{-1})

#### Question 7

7. [5 points] Show that the ridge regression estimator can be computed as least squares applied to an augmented data set. Take the original data $(\boldsymbol{Y}, \boldsymbol{X})$. Add $p$ 0's to $\boldsymbol{Y}$ and $p$ rows of $\sqrt{\lambda} \boldsymbol{I}_p$ to $\boldsymbol{X}$, apply least squares, and show that this equals $\hat{\beta}_{\text {ridge }}$.

_Answer_:


The OLS and Ridge regression estimators are given by:

$$
\hat{\beta}_{\text{ols}} = (X'X)^{-1}X'y
$$

$$
\hat{\beta}_{\text{ridge}} = (X'X + \lambda I_p)^{-1}X'y
$$


We define the augmented system:

$$
\tilde{y} = \tilde{X}\beta + e
$$

where

$$
\tilde{y} = \begin{bmatrix} y_{n \times 1} \\ 0_{p \times 1} \end{bmatrix}, \quad
\tilde{X} = \begin{bmatrix} X_{n \times p} \\ \sqrt{\lambda} I_p \end{bmatrix}, \quad
e = \begin{bmatrix} e_{n \times 1} \\ 0_{p \times 1} \end{bmatrix}
$$


The estimator $\hat{\beta}^*$ is given by:

$$
\hat{\beta}^* = (\tilde{X}'\tilde{X})^{-1} \tilde{X}' \tilde{y}
$$

Expanding:

$$
\tilde{X}'\tilde{X} = \begin{bmatrix} X' & \sqrt{\lambda} I_p \end{bmatrix} \begin{bmatrix} X \\ \sqrt{\lambda} I_p \end{bmatrix} = X'X + \lambda I_p
$$

$$
\tilde{X}' \tilde{y} = \begin{bmatrix} X' & \sqrt{\lambda} I_p \end{bmatrix} \begin{bmatrix} y \\ 0 \end{bmatrix} = X'y
$$

Thus:

$$
\hat{\beta}^* = (X'X + \lambda I_p)^{-1} X'y
$$

 Conclusion:

$$
\hat{\beta}^* = \hat{\beta^*}_{\text{ols}} = \hat{\beta}_{\text{ridge}}
$$
