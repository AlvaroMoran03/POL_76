### Cálculo de $ \hat{\beta} $

La expresión inicial es:

$$
\beta = (X'X)^{-1}(X'y)
$$


Para $ \hat{\beta} $, expandimos:

$$
\hat{\beta} = (X'X)^{-1}(X'(X\beta + e))
$$

Desarrollamos:

$$
\hat{\beta} = (X'X)^{-1}\left[X'X\beta + X'e\right]
$$

Reorganizamos:

$$
\hat{\beta} = (X'X)^{-1}(X'X)\beta + (X'X)^{-1}(X'e)
$$

Finalmente:

$$
\hat{\beta} = \beta + (X'X)^{-1}(X'e)
$$



La diferencia entre $ \hat{\beta}$  y $\beta $ es:

$$
\hat{\beta} - \beta = (X'X)^{-1}(X'e)
$$



Para el sesgo esperado, tenemos:

$$
\hat{\beta} - \mathbb{E}[\hat{\beta} | X] = (X'X)^{-1}(X'e)
$$




Partimos de la definición:

$$
\text{var}(\hat{\beta} | X) = \mathbb{E}\left[ (\hat{\beta} - \mathbb{E}[\hat{\beta} | X])(\hat{\beta} - \mathbb{E}[\hat{\beta} | X])' \middle| X \right]
$$

Sustituimos la expresión de \( \hat{\beta} - \mathbb{E}[\hat{\beta} | X] \):

$$
\text{var}(\hat{\beta} | X) = \mathbb{E}\left[ \left[ (X'X)^{-1} X'e \right] \left[ (X'X)^{-1} X'e \right]' \middle| X \right]
$$


$$
\text{var}(\hat{\beta} | X) = (X'X)^{-1} X' \mathbb{E}[ee' | X] X (X'X)^{-1}
$$

Donde:

-  $\mathbb{E}[ee' | X] $ es la matriz de varianzas y covarianzas de los errores, representada como $ \Omega $.

Finalmente:

$$
\text{var}(\hat{\beta} | X) = (X'X)^{-1} X' \Omega X (X'X)^{-1}
$$


**Nota:** No se asume homocedasticidad.
