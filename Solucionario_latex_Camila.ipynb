Número correctamente clasificado $(n) = \text{clasificados con } Y = 0 + \text{clasificados con } Y = 1$

Esto se expresa como

$$
\text{Clasificados}(Y = 0) = n(1 - \bar{y}) \hat{q_0}
$$
$$
\text{Clasificados}(Y = 1) = n(\bar{y}) \hat{q_1}
$$

Entonces, el porcentaje total correctamente predicho es:

$$
\hat{p} = \frac{n(1 - \bar{y}) \hat{q_0} + n(\bar{y}) \hat{q_1}}{n}
$$

Sustituimos los valores dados:

$$
\bar{y} = 0.70
$$

$$
\hat{q_0} = 80\%, \text{ porcentaje correctamente predicho para } Y = 0
$$

$$
\hat{q_1} = 40\%, \text{ porcentaje correctamente predicho para } Y = 1
$$

Por lo tanto,

$$
\hat{p} = (1 - \bar{y}) \hat{q_0} + \bar{y} \hat{q_1}
$$

$$
\hat{p} = (0.3)(0.8) + (0.7)(0.4)
$$

$$
\hat{p} = 0.52 \approx 52\%

$$
X' \beta + e > 0
$$

$$
P(X) = \Pr[e > -X' \beta \mid X] = 1 - G(-X' \beta) = G(X' \beta)
$$

Si el modelo índice es $$\Pr[Y = 1 \mid X] = G(X' \beta)$$, por lo que la función de probabilidad condicional es:

$$
\Pi(Y \mid X) = G(X' \beta)^y G(-X' \beta)^{1 - y} = G(Z' \beta)
$$

$$
Z = \left\{
\begin{array}{cc}
X, & Y = 1 \\
-X, & Y = 0
\end{array}
\right\}
$$

Por lo que el modelo Probit quedaría:

$$
\ell^{Probit}(\beta) = E[\log \Phi(Z' \beta)]
$$

$$
\ell_n^{Probit}(\beta) = n^{-1} \sum_{i=1}^{n} \log \Phi(Z_i' \beta)
$$

$$
\hat{\beta}_{(emily)}^{Probit} = \arg\max_{\beta} \left\{ n^{-1} \sum_{i=1}^{n} \log \Phi(Z_i' \beta) \right\}
$$

En el caso de Jacob, mantiene la muestra pero cambian los indicadores de decisión.

$$
Y^{*} = X' \beta + e

$$

\begin{array}{cc}
1, & Y^{*} < 0 \\

\end{array}
\right\}
$$

$$
X' \beta + e < 0
$$

$$
P(X) = \Pr[e < -X' \beta \mid X] = G(-X' \beta)
$$

El modelo índice es $$\Pr[Y = 1 \mid X] = G(-X' \beta)$$ por lo que la función de probabilidad condicional es:

$$
\Pi(Y \mid X) = G(-X' \beta)^y G(X' \beta)^{1 - y} = G(-Z' \beta)
$$

$$
Z = \left\{
\begin{array}{cc}
-X, & Y = 1 \\
X, & Y = 0
\end{array}
\right\}
$$

Por lo que el modelo Probit sería:

$$
\ell^{Probit}(\beta) = E[\log \Phi(-Z' \beta)]
$$

$$

$$

$$
\hat{\beta}_{(Jacob)}^{Probit} = \arg\max_{\beta} \left\{ n^{-1} \sum_{i=1}^{n} \log \Phi(-Z_i' \beta) \right\}
$$

\tilde{\beta} = \beta + \left( \frac{1}{n} \sum_{i=1}^{n} X_i X_i' \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \right)^{-1} \left( \frac{1}{n} \sum_{i=1}^{n} X_i e_i \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \right)

Por Smooth Function:

$$
h(c) = 1 \left\{ |X| \leq c \right\}
$$

$$
\mu = E[h(c)] = E\left[ 1 \left\{ |X| \leq c \right\} \right] = \int 1 \left\{ |X| \leq c \right\} dF(X) = F(c)
$$

$$
\hat{F(c)} = \hat{\mu} = n^{-1} \sum_{i=1}^{n} 1 \left\{ |X_i| \leq c \right\}
$$

La condición es:

$$
\mathbb{E}\left[ 1 \left\{ |X| \leq c \right\} \right] \leq 1 < \infty,
$$

por lo que podemos concluir, por el teorema, que el estimador es consistente al converger en probabilidad:

$$
\hat{\mu} \xrightarrow{p} \mu
$$

$$
\hat{F(c)} = \hat{\mu} = n^{-1} \sum_{i=1}^{n} 1 \left\{ |X_i| \leq c \right\} \xrightarrow{p} F(c)
$$

Además, por el método plug-in, converge en probabilidad:

$$
\frac{1}{n} \sum_{i=1}^{n} X_i X_i' \xrightarrow{p} E[XX']
$$

Entonces, por el Teorema del Mapeo Continuo, converge en probabilidad:

$$
\frac{1}{n} \sum_{i=1}^{n} X_i X_i' 1 \left\{ |X_i| \leq c \right\} \xrightarrow{p} E[XX'] F(c)
$$

Por lo que quedaría:

$$
\tilde{\beta} = \beta + \left( \frac{1}{n} \sum_{i=1}^{n} X_i X_i' \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \right)^{-1} \left( \frac{1}{n} \sum_{i=1}^{n} X_i e_i \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \right) \xrightarrow{p} \beta + \left( E[XX'] F(c) \right)^{-1} \left( E[X e] F(c) \right)
$$

Por la ley de expectativas iteradas y dado que se cumple la condición $E[e \mid X] = 0$:

$$
E\left[ E[X e \mid X] \right] = E\left[ X E[e \mid X] \right] = 0
$$

$$
\tilde{\beta} = \beta + \left( \frac{1}{n} \sum_{i=1}^{n} X_i X_i' \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \right)^{-1} \left( \frac{1}{n} \sum_{i=1}^{n} X_i e_i \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \right) \xrightarrow{p} \beta + 0
$$

**b)** Para encontrar la aproximación asintótica normal de la distribución del estimador de mínimos cuadrados ordinarios:

$$
\sqrt{n} \left( \tilde{\beta} - \beta \right) = \left( \frac{1}{n} \sum_{i=1}^{n} X_i X_i' \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \right)^{-1} \left( \frac{1}{\sqrt{n}} \sum_{i=1}^{n} X_i e_i \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \right)
$$

Donde si $(X_i e_i)$ es i.i.d. y tiene una media cero, se define:

$$
\varOmega = E\left[ X X' e^2 \right]
$$

Además, el CLT requiere que $\varOmega < \infty$.

Dado que el primer elemento converge en probabilidad:

$$
\left( \frac{1}{n} \sum_{i=1}^{n} X_i X_i' \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \right)^{-1} \xrightarrow{p} \left( E[XX'] F(c) \right)^{-1}
$$

Mientras que el segundo converge en distribución:

$$
\frac{1}{\sqrt{n}} \sum_{i=1}^{n} X_i e_i \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \xrightarrow{d} N\left( 0, \varOmega F(c) \right)
$$

Por el teorema de Slutsky, donde un elemento converge en probabilidad y el segundo en distribución, mantiene la distribución de probabilidad:

$$
\sqrt{n} \left( \tilde{\beta} - \beta \right) = \left( \frac{1}{n} \sum_{i=1}^{n} X_i X_i' \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \right)^{-1} \left( \frac{1}{\sqrt{n}} \sum_{i=1}^{n} X_i e_i \mathbb{\mathbf{1}}_{\left\{ |x_i| < \mathbf{c} \right\}} \right) \xrightarrow{d} \left( E[XX'] F(c) \right)^{-1} N\left( 0, \varOmega F(c) \right)
$$
