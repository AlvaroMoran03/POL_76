# Multinomial Response

1. **Pregunta 1**

    (a) For estimating the mean of a nonnegative random variable $y$ using a random quasi-log likelihood for a random draw is $\gamma_i(\mu) = y_i \log(\mu) - \mu, \mu > 0$
      (where terms not depending on $\mu$ have been dropped). Letting $\mu_0 \equiv E(y_i)$, we have
      $$E[l_i(u)] = \mu_0 \log(\mu) - \mu$$
      Show that this function is uniquely maximized at $\mu = \mu_0$.

   (b) The gamma (exponential) quasi-log likelihood is $$l_i(\mu) = -\frac{y_i}{\mu} - \log(\mu), \quad \mu > 0.$$ Show that this function is uniquely maximized at $\mu = \mu_0$.

   **Solución**:

   - **(a)** Maximizando:

     $$\frac{\partial \gamma_i(\mu)}{\partial \mu} = \frac{y_i}{\mu} - 1 = 0$$
     de donde se obtiene $y_i = \mu$. Luego,
     $$E(y_i) = E(\mu) \quad \Rightarrow \quad \mu_0 = \mu.$$
     Por lo tanto, la función de cuasi-verosimilitud es máxima cuando $\mu = \mu_0$.

     Derivando la función de verosimilitud:
     $$\frac{\partial E[l_i(\mu)]}{\partial \mu} = \frac{\mu_0}{\mu} - 1 = 0 \quad \Rightarrow \quad \mu = \mu_0.$$
     De nuevo, esto confirma que la función se maximiza en $\mu = \mu_0$.

   - **(b)** Aplicando valor esperado:

     $$l_i(\mu) = -\frac{y_i}{\mu} - \log(\mu)$$
     $$\mathbb{E}[l_i(\mu)] = -\frac{1}{\mu} \mathbb{E}(y_i) - \log(\mu)$$

     Maximizando la función de verosimilitud:

     $$\frac{\partial \mathbb{E}[l_i(\mu)]}{\partial \mu} = \mu^{-2} \mathbb{E}(y_i) - \mu^{-1} = 0$$

     Suponiendo que $\mathbb{E}(y_i) = \mu_0$, como en (1a):

     $$\mu^{-2} \mu_0 - \mu^{-1} = 0 \quad \therefore \quad \mu = \mu_0$$

     Por tanto, $\mu_0$ es el valor que maximiza $l_i(\mu)$.

2. **Pregunta 2**

   In a balanced panel data setting, i.e. $\bigl\{\{y_{it}, x_{it}\}_{t=1}^{T}\bigr\}_{i=1}^n,$
   consider an unobserved effects model for count data with exponential regression function:

   $$\mathbb{E}(y_{it} \mid x_{i1}, \ldots, x_{iT}, c_i) = c_i \exp(x_{ii}\beta) \tag{1}$$

   If
   $\mathbb{E}(c_i \mid x_{i1}, \ldots, x_{iT}) = \exp(\alpha + \bar{x}_i \gamma)$,
   find $\mathbb{E}(y_{it} \mid x_{i1}, \ldots, x_{iT})$
   where $\bar{x}_i = T^{-1}(x_{i1} + \cdots + x_{iT})$.
   
   **Solución**:

   Aplicando valor esperado:
   $$\mathbb{E}\!\Big[\mathbb{E}(y_{it} \mid x_{i1}, \ldots, x_{iT}, c_i) \;\Big\vert\; x_{i1}, \ldots, x_{iT}\Big] = \mathbb{E}\!\Big[c_i \exp(x_{ii}\beta) \;\Big\vert\; x_{i1}, \ldots, x_{iT}\Big]$$
  Usando la Ley de Expectativas Iteradas:
   $$\mathbb{E}(y_{it} \mid x_{i1}, \ldots, x_{iT}) = \exp(x_{ii}\beta)\;\mathbb{E}(c_i \mid x_{i1}, \ldots, x_{iT})$$

   $$\mathbb{E}(y_{it} \mid x_{i1}, \ldots, x_{iT}) = \exp\bigl(x_{ii}\beta \bigr)\exp\bigl(\alpha + \bar{x}_i\gamma\bigr) $$

   $$\therefore \mathbb{E}(y_{it} \mid x_{i1}, \ldots, x_{iT}) = \exp\bigl(x_{ii}\beta + \alpha + \bar{x}_i\gamma\bigr)$$

3. **Pregunta 3**

   Let $\text{patents}$ be the number of patents applied for by a firm during a given year. Assume that the conditional expectation of $\text{patents}$ given $\text{sales}$ and $\text{RD}$ is

   $$\mathbb{E}(\text{patents} \mid \text{sales}, RD) = \exp\bigl[\beta_0 + \beta_1 \log(\text{sales}) + \beta_2 RD + \beta_3 RD^2\bigr],$$

   where *sales* is annual firm sales and *RD* is total spending on research and development over the past 10 years.

   (a) How would you estimate the $\beta_j$? Justify your answer by discussing the nature of patents.

   (b) How do you interpret $\beta_1$?

   (c) Find the partial effect of *RD* on $\mathbb{E}(\text{patents} \mid \text{sales}, RD)$.

   **Solución**:

   - **(a)** Dado que el número de patentes es una variable de conteo, usaría una regresión Poisson. Empleamos Poisson cuando la variable dependiente es un conteo. La empresa puede solicitar un número indeterminado de patentes en un año, por lo que
     $$\text{patents} \in \mathbb{N} = \{0, 1, 2, 3, \ldots\} $$
     *(Nota: No hay un orden implícito entre estos valores, por lo que no es variable categórica.)*

   - **(b)** Cuando las ventas anuales (*sales*) aumentan en 1%, el número de patentes que solicita la empresa aumenta en $\beta_1 / 100$.

   - **(c)** El efecto marginal de *RD* sobre $\mathbb{E}(\text{patents} \mid \text{sales}, RD)$ es:

     $$\frac{\partial \mathbb{E}(\text{patents} \mid \text{sales}, RD)}{\partial RD} = (\beta_2 + 2\beta_3 RD)\exp\bigl(\beta_0 + \beta_1 \log(\text{sales}) + \beta_2 RD + \beta_3 RD^2\bigr)$$
